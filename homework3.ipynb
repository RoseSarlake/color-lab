```python
# -*- coding: utf-8 -*-
"""
Created on Wed Sep 29 12:51:09 2021

@author: Linna
"""
from CSL_homework_1 import read_envi
from CSL_homework_2 import *
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir("ColorChecker_Specim_V10E\capture")

# read the spectral cube
sample, wavelengths = read_envi('colorchecker_sample_0007')
white = read_envi('WHITEREF_colorchecker_sample_0007')[0]
dark = read_envi('DARKREF_colorchecker_sample_0007')[0]

shapeSample = sample.shape
shapeWhite  = white.shape
shapeDark   = dark.shape

white = white.mean(axis=0).reshape((1,shapeWhite[1],shapeWhite[2])).repeat(shapeSample[0], axis=0)  
dark = dark.mean(axis=0).reshape((1,shapeDark[1],shapeDark[2])).repeat(shapeSample[0],axis=0)

refCube = (sample-dark)/(white-dark)

# Test the effect of different standard illuminants (‘A’, ‘D65’, ‘F11’, etc.) on the ColorChecker spectral image
rgbImg_A = spim2rgb(refCube, wavelengths, lsource='A')
rgbImg_d65 = spim2rgb(refCube, wavelengths, lsource='D65')
rgbImg_f11 = spim2rgb(refCube, wavelengths, lsource='F11')
rgbImg_l2 = spim2rgb(refCube, wavelengths, lsource='L2')
rgbImg_d50 = spim2rgb(refCube, wavelengths, lsource='D50')
rgbImg_f8 = spim2rgb(refCube, wavelengths, lsource='F8')

plt.figure()
#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(3,2) 

axarr[0,0].imshow(rgbImg_A)
axarr[0,0].set_title("A")
axarr[0,1].imshow(rgbImg_d65)
axarr[0,1].set_title("D65")
axarr[1,0].imshow(rgbImg_f11)
axarr[1,0].set_title("F11")
axarr[1,1].imshow(rgbImg_l2)
axarr[1,1].set_title("L2")
axarr[2,0].imshow(rgbImg_d50)
axarr[2,0].set_title("D50")
axarr[2,1].imshow(rgbImg_f8)
axarr[2,1].set_title("F8")


# Select reflectance spectra from two color patches (e.g., red and green). 
# Compare the spectra and design an emission spectrum which maximizes or minimizes the color 
# difference (use visual evaluation) between the two selected color patches in an RGB image

    #check the spectra o two color patches
red = refCube[200,600,:]
green = refCube[200,700,:]
plt.plot(wavelengths, red, color='r')
plt.plot(wavelengths, green, color='g')

    # to maximize the color difference, need to "enhance" the most different spectra range
spc_max = np.zeros(len(wavelengths))
spc_max[45:50] = 1
spc_max[65:75] = 1
plt.plot(spc_max)
rgbImg_max = spim2rgb(refCube, wavelengths, lsource= spc_max)
plt.imshow(rgbImg_max)
    # to minimize the color difference, need to find the identical spectra
spc_min = np.zeros(len(wavelengths))
spc_min[55:57] = 1
spc_min[105:110] = 1
plt.plot(spc_min)
rgbImg_min = spim2rgb(refCube, wavelengths, lsource= spc_min)
plt.imshow(rgbImg_min)

# try another pair
pink = refCube[350,550,:]
blue = refCube[500,550,:]
plt.plot(wavelengths, pink, color='r')
plt.plot(wavelengths, blue, color='b')

spc_max0 = np.zeros(len(wavelengths))
spc_max0[25:27] = 1
plt.plot(spc_max0)
rgbImg_max0 = spim2rgb(refCube, wavelengths, lsource= spc_max0)
plt.imshow(rgbImg_max0)

spc_min0 = np.zeros(len(wavelengths))
spc_min0[52:54] = 1
plt.plot(spc_min0)
rgbImg_min0 = spim2rgb(refCube, wavelengths, lsource= spc_min0)
plt.imshow(rgbImg_min0)
```
